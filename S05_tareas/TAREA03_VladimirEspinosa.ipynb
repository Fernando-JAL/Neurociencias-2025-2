# PyTorch
PyTorch es un marco de aprendizaje automático de código abierto que acelera el camino desde la creación de prototipos de investigación hasta la 

### Características principales
Framework desarrollado por Facebook.

Usa ejecución dinámica (define-by-run).

Muy intuitivo y cercano a NumPy.

Integra bien con Python y permite depuración con herramientas estándar.

Compatible con GPU, CPU (y parcialmente con TPU).

Tiene herramientas como TorchVision, TorchAudio, TorchServe, TorchScript.

### Ventajas
Fácil de usar y depurar: ideal para aprender y hacer prototipos rápidos.

Muy popular en investigación científica y papers.

Código limpio, pythonic y flexible.

Gran integración con librerías de Python: como NumPy, SciPy, scikit-learn.

Activa comunidad de desarrolladores académicos.

Compatibilidad creciente con herramientas de producción (TorchServe, ONNX, etc).

### Desventajas
Ecosistema de producción aún en desarrollo (aunque ha mejorado mucho).

Menor soporte nativo para móviles y navegadores.

Distribución y paralelismo requieren configuración manual más detallada.

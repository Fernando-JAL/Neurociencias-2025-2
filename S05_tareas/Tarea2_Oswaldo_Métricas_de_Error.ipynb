1. MAE – Error Absoluto Medio (Mean Absolute Error)
Es el promedio de los errores absolutos entre las predicciones del modelo y los valores reales.
¿Cuánto se equivoca el modelo en promedio, sin importar si se pasa o se queda corto?
2. MSE – Error Cuadrático Medio (Mean Squared Error)
Es el promedio del cuadrado de los errores.
Le da más peso a los errores grandes. Si tu modelo se equivoca mucho en una predicción, lo penaliza mucho más.
3. RMSE – Raíz del Error Cuadrático Medio (Root Mean Squared Error)
Es la raíz cuadrada del MSE. También penaliza errores grandes, pero vuelve la métrica interpretable porque regresa a las unidades originales (como pesos, días, grados, etc.).
5. R² – Coeficiente de determinación
Mide cuánta proporción de la variabilidad de los datos reales es explicada por el modelo.
Si R² = 0.9 → mi modelo explica el 90% de la variabilidad.
Si R² = 0.0 → mi modelo no explica nada mejor que adivinar con el promedio.
Puede tbm ser negativo si el modelo es muy malo.


```python
# Métricas de error en regresión: MSE, MAE, RMSE, R²

import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# EJEMPLO
# Valores reales
y_true = np.array([3.0, -0.5, 2.0, 7.0])

# Valores previstos por el modelo
y_pred = np.array([2.5, 0.0, 2.1, 7.8])

# Fórmula: MSE = (1/n) * Σ(y_i - ŷ_i)^2
mse = mean_squared_error(y_true, y_pred)

# Fórmula: MAE = (1/n) * Σ|y_i - ŷ_i|
mae = mean_absolute_error(y_true, y_pred)

# Fórmula: RMSE = √MSE
rmse = np.sqrt(mse)

# Fórmula: R² = 1 - (SS_res / SS_tot)
r2 = r2_score(y_true, y_pred)

# Mostrar resultados
print("Resultados de las métricas de error:")
print(f"MSE (Mean Squared Error): {mse:.4f}")
print(f"MAE (Mean Absolute Error): {mae:.4f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.4f}")
print(f"R² (Coeficiente de determinación): {r2:.4f}")

# Diagrama visual
plt.figure(figsize=(6, 4))
plt.plot(y_true, label="Valor real", marker="o")
plt.plot(y_pred, label="Predicción", marker="x")
plt.title("Comparación entre valores reales y predichos")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```

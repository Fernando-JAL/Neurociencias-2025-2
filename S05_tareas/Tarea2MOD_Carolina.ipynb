En los problemas de regresión, se utilizan diversas métricas estadísticas para evaluar el rendimiento de los modelos predictivos. Estas métricas tienen como objetivo principal cuantificar la diferencia entre los valores reales observados y los valores predichos por el modelo. Al medir esta discrepancia, es posible determinar qué tan preciso es el modelo en sus predicciones. Algunas de las métricas más comunes incluyen el error cuadrático medio (MSE), el error absoluto medio (MAE) y el coeficiente de determinación (R²), cada una con sus propias ventajas y limitaciones según el tipo de datos y el contexto del análisis. Al conocer bien éstas métricas es posible saber si nuestro modelo es eficiente, necesita ser ajustado o si debe considerarse una alternativa más adecuada para mejorar la capacidad predictiva.

**MSE (MEAN SQUARED ERROR)**

* *Definición*: Es una métrica que se utiliza para cuantificar cuán precisas son las predicciones de un modelo estadístico o de regresión. Específicamente, el MSE calcula el promedio de las diferencias al cuadrado entre los valores observados (reales) y los valores predichos por el modelo.
Este valor siempre será cero o positivo. Si el modelo predice perfectamente todos los valores, el MSE será 0, lo cual indica ausencia total de error.
A medida que las predicciones se alejan de los valores reales, el MSE se incrementa, lo que refleja un mayor nivel de error en el modelo.

* *Aplicación*: El MSE se utiliza para medir la desviación entre las predicciones de un modelo y los valores reales, siendo un valor bajo indicativo de mayor precisión. En el contexto de la regresión, refleja el promedio de los errores al cuadrado, otorgando más peso a los errores más grandes. Además, permite comparar distintos modelos y seleccionar el que presente el menor error. Esta métrica se aplica de manera amplia en diversos campos tales como los modelos de regresión, el entrenamiento de algoritmos en machine learning, los pronósticos de series temporales, el procesamiento de imágenes y señales, y el modelado en ciencias experimentales.

Es ampliamente aplicado en modelos de regresión, entrenamiento de machine learning, y en el análisis de series temporales, como en la predicción de ventas o pronósticos meteorológicos.

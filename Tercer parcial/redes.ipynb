### Cargar datos 

```python
path = r"/Users/sabrina/Documents/UNI/A-NEUROCIENCIAS/SEXTO/MODELOS2/Neurociencias-2025-2/S03_datasets/iris/iris.csv"
df = pd.read_csv(path)
df.head(3)
```

```python
X_, y = df.iloc[:, :-1], df.iloc[:, -1].astype('category').cat.codes # con cat.codes segun yo las etiquetas las pasas a numeros 
# estas usando species como tu variable 

# normalizamos los datos
X = Scaler().fit_transform(X_)

# dividimos los datos en train y test
y_cat = to_categorical(y) # toma el vector y lo convierte a una matriz de (150,3) 
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size = 0.2, random_state = 0) 
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential() # es el esqueleto de la red neuronal, es decir vamos a crear la red neuronal !!

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 10, activation = 'relu', input_shape=(4,))) # usamos re lu para que no den valores negativos 
# 10 es la capa oculta y 4 es la de entrada por sepal y petal length y width
# el 10 es por una regla de dedo multiplicas 4x3 y 4x2 y es 12 y 8 y puedes escoger cualquier numero entre esos dos

# Adding the output layer
classifier.add(Dense(units = 3, activation = 'softmax')) # softmax te esta dando la probabilidad de que pertenezca a cada clase y la suma es 1
# para problemas de clasificacion es mejor usar softmax y para problemas de regresion es mejor usar linear


# Compiling the ANN es cuando pasa cada epoch 
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']) # categorical_crossentropy es para problemas de clasificacion y binary_crossentropy es para problemas de regresion
# la de categorical es la funcion que te calcula el errror 
# adam es la que se va a la derivada 

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, epochs=150, validation_data=(X_test, y_test))
```

```python
# evaluamos el modelo 
y_pred = classifier.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

print("\nğŸ” Reporte de clasificaciÃ³n:")
print(classification_report(y_true, y_pred_classes, target_names=df.Species.unique()))
```

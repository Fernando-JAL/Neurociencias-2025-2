1. Dataset de Coactivation matrix
    - Utilizar el método de codo para determinar el número de cluster óptimo (adecuado)


```python
# Cargar los datos
mat_path = r"..\..\S03_datasets\coactivation_matrix\Coactivation_matrix.mat"
mat_data = scipy.io.loadmat(mat_path)

# Obtener la matriz de coactivación y las coordenadas
coactivation_matrix = mat_data['Coactivation_matrix']
coordinates = mat_data['Coord']
```

```python
coactivation_matrix.shape
```

```python
coordinates.shape
```

```python
# Normalizando datos
X_std = StandardScaler().fit_transform(coactivation_matrix)
```

# K means

```python
# Método del codo
inertia = []
for no_clusters in range(2, 15):
    kmeans = KMeans(n_clusters=no_clusters)
    kmeans.fit(coactivation_matrix)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 15), inertia, '.-')

plt.xlabel('número de clusters')
plt.ylabel('valor de inertia')
plt.title('Método del codo')
```

```python
no_clusters = 5
kmeans = KMeans(n_clusters=no_clusters).fit(coactivation_matrix)
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

x, y, z = coordinates[:, 0], coordinates[:, 1], coordinates[:, 2]
# Creating plot
ax.scatter3D(x, y, z, c=kmeans.labels_)
```

```python
# Se lee el dataset
path = r"..\..\S03_datasets\Brain_tumor"
bt_path = os.path.join(path, 'Brain Tumor.csv')

# Se construye el dataframe
bt_df = pd.read_csv(bt_path)
```

```python
bt_df.head()
```

```python
# Se define el conjunto de datos
X = bt_df.iloc[:, 2:-1]

# Normalizando datos
X_std = StandardScaler().fit_transform(X)
```

```python
# metodo del codo
# Método del codo
inertia = []
for no_clusters in range(2, 15):
    kmeans = KMeans(n_clusters=no_clusters)
    kmeans.fit(X_std)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(2, 15), inertia, '.-')

plt.xlabel('número de clusters')
plt.ylabel('valor de inertia')
plt.title('Método del codo')
```

```python
# clustering jerárquico
Z = linkage(X_std, 'ward')

plt.figure(figsize=(15, 5))
ddata = dendrogram(Z)
for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
    x = 0.5 * sum(i[1:3])
    y = d[1]
    plt.plot(x,y,'o',c=c)
    plt.annotate('%.3g'%y, (x,y), xytext=(0,-5), 
                 textcoords="offset points", va="top", ha="center")

plt.plot()
```

```python
# kmeans
kmeans = KMeans(n_clusters=2).fit(X_std)
kmeans_labels = kmeans.labels_

# gaussian mixture
gm = GaussianMixture(n_components=2, random_state=0).fit(X_std)
gm_labels = gm.predict(X_std)


# clustering jerárquico
Z = linkage(X_std, 'ward')
clusters = fcluster(Z, t=120, criterion='distance')
```

```python
cm_gauss = confusion_matrix(bt_df.Class, gm_labels)
cm_kmeans = confusion_matrix(bt_df.Class, kmeans_labels)
cm_cluster = confusion_matrix(bt_df.Class, clusters)

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

sns.heatmap(cm_kmeans, annot=True,
           cmap="Blues", ax = axes[0]).set_title("Matriz de confusión de kmeans")

axes[0].set_xlabel('Predicción')
axes[0].set_ylabel("Real")

sns.heatmap(cm_gauss, annot=True,
           cmap="Blues", ax = axes[1]).set_title("Matriz de confusión de Gaussian mixture")

axes[1].set_xlabel('Predicción')
axes[1].set_ylabel("Real")


sns.heatmap(cm_gauss, annot=True,
           cmap="Blues", ax = axes[2]).set_title("Matriz de confusión de Cluster jerarquico")

axes[2].set_xlabel('Predicción')
axes[2].set_ylabel("Real")
```


3. De su previa investigación:
    - qué es precision?
    - qué es accuracy?
    - qué es recall?
    - qué es f1-score?


accuracy - mide el porcentaje de predicciones correctas sobre el total de ejemplos

precision - cuenta todas las veces que el modelo predijo "positivo"

recall - de todos los casos realmente positivos, ¿cuántos detectó el modelo?

f1-score - mide el equilibrio entre precisión y recall, es la media armónica

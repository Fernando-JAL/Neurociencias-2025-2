# Iris Dataset

```python
# Importar paqueterías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Cargar datos y crear dataframe
path = r"C:\Users\clau\Desktop\Repo-2025\Neurociencias-2025-2\S03_datasets\iris\iris.csv"
```

```python
df = pd.read_csv(path)
df.head() # Primeros datos

# Descartar columna Species
remove = df.drop(columns=['Species'])
```

```python
# Cargar dataset desde sklearn
iris = datasets.load_iris()
X_iris = iris.data  # Variables numéricas
y_iris = iris.target  # Etiquetas de clase (no se usan en PCA)

# 1. Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_iris)
```

```python
# 2. Aplicar PCA con 3 componentes principales
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
# 3. Varianza explicada
var_ratio = pca.explained_variance_ratio_
total_var = np.sum(var_ratio)
print("Varianza por cada componente:", var_ratio)
print("Varianza total explicada:", total_var)
```

```python
# 4. Gráfico 3D con los datos proyectados
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
scatter = ax.scatter(
    X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y_iris, cmap="viridis", edgecolor="k"
)
ax.set_title("PCA: Iris Dataset (3 Componentes)")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.colorbar(scatter, ax=ax, label="Clase")
plt.show()
```

# Brain Tumor Dataset

```python
# Importar paqueterias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar dataset
path_bt = r"C:\Users\clau\Desktop\Repo-2025\Brain Tumor.csv"
df = pd.read_csv(path_bt) # Definir path
df.head() # Primeros valores
```

```python
# Eliminar las columnas 'Image' y 'Class'
df_newbt = df.drop(columns=["Image", "Class"])
df_newbt
```

```python
# La última columna "Coarseness" muestra los mismos valores (también aparece 0 en la desviación estándar de esa columna al normalizar, indicando que es una
# constante en todos los datos).
```

```python
# Normalizar datos
scaler = StandardScaler()
scaled_bt = scaler.fit_transform(df_newbt)

# Verificar
print("Media post-normalización:", np.mean(scaled_bt, axis=0)) # Media de cada columna debe ser 0 aprox.
print("Desviación estándar post-normalización:", np.std(scaled_bt, axis=0)) # Desviación estándar de cada columna debe ser 1 aprox.
```

```python
# Aplicar PCA
pca = PCA()
pca.fit(scaled_bt)
```

```python
# Varianza explicada por cada componente
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
# Determinar el número de componentes necesarios para el 90% de la varianza
ac_var = np.cumsum(explained_variance) # Varianza acumulada
comp_90 = np.argmax(ac_var >= 0.90) + 1
print(f"El no. de componentes necesarios para conservar el 90% de la varianza es: {comp_90}")
```

```python
# Reducción de Dimensionalidad a 4 componentes
pca_4 = PCA(n_components=4)
df_pca4 = pca_4.fit_transform(scaled_bt)
df_pca4
```

```python
# Otro método
# Truncated SVD (más eficiente para datos normalizados)
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=comp_90)
df_svd = svd.fit_transform(scaled_bt)
```

```python
# Comparar ambos métodos
var_pca = np.sum(pca_4.explained_variance_ratio_)
var_svd = np.sum(svd.explained_variance_ratio_)

print(f"Método 1: {var_pca}")
print(f"Método 2: {var_svd}")
```

```python
# Graficar Datos Originales
df_sub = pd.DataFrame(scaled_bt[:, :4], columns=df_newbt.columns[:4]) # Otro data frame
df_1pca = pd.DataFrame(df_pca4, columns=[f"PC{i+1}" for i in range(comp_90)])

sns.pairplot(df_sub, plot_kws={"color": "black"}, diag_kws={"color": "grey"})
plt.suptitle("Pairplot: Datos Originales (Primeras 4 Variables)", y=1.02)
plt.show()
```

```python
# Graficar: Datos con PCA Aplicado
sns.pairplot(df_1pca, plot_kws={"color": "black"}, diag_kws={"color": "grey"})
plt.suptitle("Pairplot: Datos con PCA Aplicado", y=1.02)
plt.show()
```

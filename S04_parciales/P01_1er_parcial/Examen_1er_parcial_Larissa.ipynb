# Pt. 1

### Reproducir el análisis de PCA aplicado al Iris dataset, utilizando 3 componentesdos

```python
path = r"C:\Users\laris\OneDrive\Documentos\LNC\Modeloscomputacionales\Modelos Computacionales II\S03_datasets\iris\iris.csv"
```

```python
df_iris = pd.read_csv(path)
df_iris.head()
X_iris = df_iris.iloc[:, :-1]
X_iris
```

```python
media, sigma = X_iris.mean(axis=0), X_iris.std(axis=0)
X_iris_std = (X_iris - media)/sigma
```

```python
pca_iris = PCA(n_components=3)
X_iris_pca = pca_iris.fit_transform(X_iris_std)
```

### 1.1 Indicar el % de la info usando los 3 componentes principales 

```python
var_exp_iris = pca_iris.explained_variance_ratio_ * 100
for i, var in enumerate(var_exp_iris, 1):
    print(f"Componente Principal {i}: {var:.2f}%")
print("Porcentaje de varianza explicada por los 3 componentes principales:", var_exp_iris.sum())
```

### 1.2 Hacer el scatter 3D de los datos proyectados, con cualquiera de los tres métodos 

```python
df_iris = pd.read_csv(path)
X = df_iris.iloc[:, :4]  
y = df_iris['Species']    

# Estandarización de datos
X_std = StandardScaler().fit_transform(X)

# Aplicar PCA con 3 componentes principales
pca = PCA(n_components=3)
data_proyectada = pca.fit_transform(X_std)

# Crear DataFrame con los datos proyectados
df_proyectado = pd.DataFrame(data_proyectada, columns=['PC1', 'PC2', 'PC3'])
df_proyectado['Species'] = y

# Mostrar varianza explicada
explained_variance = pca.explained_variance_ratio_
print(f"Varianza explicada por cada componente: {explained_variance}")
print(f"Varianza total explicada: {np.sum(explained_variance) * 100:.2f}%")


colors = {'setosa': 'blue', 'versicolor': 'green', 'virginica': 'red'}


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for species, color in colors.items():
    subset = df_proyectado[df_proyectado['Species'] == species]
    ax.scatter(subset['PC1'], subset['PC2'], subset['PC3'], 
               label=species, color=color, edgecolors='k', alpha=0.8)

ax.set_xlabel("Componente Principal 1")
ax.set_ylabel("Componente Principal 2")
ax.set_zlabel("Componente Principal 3")
ax.set_title("Análisis PCA en 3D - Iris Dataset")
ax.legend()
plt.show()
```

### Realizar el análisis de PCA utilizando el dataset Brain_tumor, archivo Brain tumor.csv

```python
df_brain = pd.read_csv(r"C:\Users\laris\OneDrive\Documentos\LNC\Modeloscomputacionales\Modelos Computacionales II\S03_datasets\Brain_tumor\Brain Tumor.csv")
```

```python
X_brain = df_brain.drop(columns=['Image', 'Class', 'Coarseness']) 
X_brain = X_brain.select_dtypes(include=[np.number])
X_brain_std = (X_brain - np.mean(X_brain, axis=0))/np.std(X_brain, axis=0)
X_brain_std
```

```python
pca_brain = PCA()
X_brain_pca = pca_brain.fit_transform(X_brain_std)
X_brain_pca
```

```python
df_brain['Coarseness'].var()
```

### Determinar el porcentaje de aporte de cada vector propio

**¿Qué noto de extraño con la columna "Coarseness" al normalizar el dataset?**


Como podemos ver en la línea de código anterior, la varianza de 'Coarseness' da 0.0, lo que quiere decir que todos sus valores son exactamemte iguales. La varianza mide cuánto se dispersan los datos alrededor de su media, si la varianza es 0, no hay dispersión , lo que implica que no hay variabilidad en los datos. Para evitar errores en nuestro código, quitamos esa columna. 

```python
var_exp_brain = pca_brain.explained_variance_ratio_ * 100

total_variance = var_exp_brain.sum()

print("Porcentaje de aporte de cada vector propio:")
for i, percentage in enumerate(var_exp_brain, start=1):
    print(f"Componente Principal {i}: {percentage:.2f}%")
    
print(f"Porcentaje total de varianza explicada: {total_variance:.2f}%")
```

### Determinar cuántos componentes principales son necesarios para obtener al menos un 90% de los datos

```python
var_exp_acumulada_brain = np.cumsum(var_exp_brain)
n_componentes = np.argmax(var_exp_acumulada_brain >= 90) + 1
print(f"Se necesitan {n_componentes} componentes principales para explicar al menos el 90% de la varianza.")
```

### Realizar la proyección de la data original al espacio reducido mediante 2 métodos y comparar resultados 

**MÉTODO 1: PCA CON SKLEARN**

```python
pca_brain = PCA(n_components=2)
X_brain_pca_sklearn = pca_brain.fit_transform(X_brain_std)
```

**MÉTODO 2:PCA MANUAL**

```python
cov_matrix_brain = np.cov(X_brain_std.T)
eig_vals_brain, eig_vectors_brain = np.linalg.eig(cov_matrix_brain)
```

```python
# Ordenar valores y vectores propios en orden descendente
sorted_indices = np.argsort(eig_vals_brain)[::-1]
eig_vals_brain = eig_vals_brain[sorted_indices]
eig_vectors_brain = eig_vectors_brain[:, sorted_indices]

# Seleccionar los dos primeros vectores propios
W_brain = eig_vectors_brain[:, :2]
X_brain_pca_manual = X_brain_std.dot(W_brain).to_numpy()

# Comparación visual
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X_brain_pca_sklearn[:, 0], X_brain_pca_sklearn[:, 1], c='r', alpha=0.5)
axes[0].set_title("PCA con sklearn")
axes[1].scatter(X_brain_pca_manual[:, 0], X_brain_pca_manual[:, 1], c='b', alpha=0.5)
axes[1].set_title("PCA manual")
plt.show()
```

### ARBOLES DE DECISION (SUPERVISADO)
* Nodo raiz (umbral petal lenght < o igual a 2.1) => Ramas (True=setosa)(False=condicion) => Nodo hojas (Condicion: si el dato esta debajo de la linea magica va a ser 'versicolor', y si no sera 'virginica').
* Tener un exceso de nodo hojas en un intento de querer hacerlo muy muy preciso podria provocar que un 'overfitting'. En que punto te detienes? A veces se necesita 'podar el arbol' (no hay que contarlo mucho para evitar el underfitting).
* Que tanta informacion me da esta variable (que tan bien esta clasificada) y que tanta esta otra y esta otra..., la variable que me da mas informacion de esta features sera mi nodo raiz (uno de baja entropia = mayor orden = mas facil de clasificar) y a partir de esa variable la computadora con el criterio que especifiquemos ('gini' 'entropy' 'log_loss') va a definir el umbral.
* Que tan homogeneo esta la base de datos, este dataset por ejemplo es muy homogeneo porque las tres clases en as que se divide tiene el mismo numero de datos 50. Sin embargo esto no siempre va a pasar, a veces tendremos dos clases una con 98 datos y otra con 30, ahi ya no es heterogeneo.
* Calcuamos la ntropia para las 4 caracteristicas (ancho y largo) y la de menos entropia sera nuestro nodo raiz, a partir de ese valor vamos a calcular la ganancia o perdida de informacion del resto de caracteristicar y definiremos nuestra nodo hoja que sera la siguiente caracteristica a partir de la cual vamos a definir.
* Si son 100 de columnas y miles de filas ya no es muy recomendable este metodo, tardaria demasiado.

* El valor de la entropia es donde haya mas desorden, si tenemos solo dos variables nuestra entropia 
probabilidad determina entropia y a partir de la entropia 
estropia para cada variable 
entre mas mezclado mas alto el valor de estrpia porque hay mucho desorden. 

```python
df.describe(include='all')
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
```

### ARBOL DE DECISION
sklearn.tree.plot_tree(decision_tree, *, max_depth=None, feature_names=None, class_names=None, label='all', filled=False, impurity=True, node_ids=False, proportion=False, rounded=False, precision=3, ax=None, fontsize=None)

```python
clf = DecisionTreeClassifier(criterion='entropy', random_state=99)
clf.fit(X, y)
```

```python
df['Species'].unique()
```

```python
plt.figure(figsize = (6, 4))
plot_tree(clf, filled = True, feature_names = df.columns[:-1], class_names = df['Species'].unique())
plt.show()
```

```python
observado = clf.predict(X)
df['clf'] = observado
```

```python
DecisionTree_cm = confusion_matrix(df.Species, df['clf'], labels=df.Species.unique())

sns.heatmap(DecisionTree_cm, annot=True,
           xticklabels = df.Species.unique(), yticklabels = df.Species.unique(),
           cmap = "Blues").set_title("Matriz de Confusion Decision Tree")

plt.xlabel('Prediccion')
plt.ylabel("Real")
```

### DIVIDIMOS EN TRAIN TEST 70:30

```python
#X = 'medias', y = 'setosa' 'versicolor' 'virginica'
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn import model_selection
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.3)
print("Shape datos entrenamiento", X_train.shape, X_test.shape, y_train.shape, y_test.shape)

clf3 = DecisionTreeClassifier(criterion='entropy', random_state=99)
clf3.fit(X_train, y_train)

y_calc = clf3.predict(X_test)
pd.crosstab(y_test, y_calc)
```

```python
print(classification_report(y_test, y_calc))
```

```python
cm7030 = confusion_matrix(y_test, y_calc)
sns.heatmap(cm7030, annot=True, cmap="Blues", xticklabels=df.Species.unique(), yticklabels=df.Species.unique())
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de confusión - LDA 70:30")
plt.show()
```

### AHORA SOBRE KFOLD

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
tree = DecisionTreeClassifier(criterion='entropy', random_state=42)
tree.fit(X,y)
```

```python
cv = KFold(n_splits = 10, shuffle = True)
```

```python
## ESTO ES CROSS VALIDATION: VA JUNTO CON EL KFOLD
scores = cross_val_score(tree, X, y, scoring = 'accuracy', cv = cv)
scores
```

```python
np.mean(scores)
```

```python
for i, (train_index, test_index) in enumerate(cv.split(X)):
    print(f"Fold {i}:")
    print(f" Train: index={train_index}")
    print(f" Test: index={test_index}")
```

### HIPERPARAMETRO: 1) DETERMINAR NUMERO DE PROFUNDIDAD ("MAXIMO DE PROFUNDIDAD")

```python
for i in range (1, 10):
    tree = DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=i)
    tree.fit(X,y)
    
    cv = KFold(n_splits=10, shuffle=True)
    scores = cross_val_score(tree, X, y, scoring="accuracy", cv=cv)
    score = np.mean(scores)
    print(f"La precision para la profundidad = {i} es {score}")
```

Si me quedara unicamente con la primera decision que es petal length obtendriamos un performance del 54%. Sin embrago, si vamos al segundo nivel tendriamos el 94% y asi sucesivamente, recuerda que lo mejor es no usar de mas por el "overfitting".

### ARBOL DE DECISION PARA REGRESION

Este ya no es un problema de clasificacion, aqui es uno de regresion, aqui lo que yo quiero saber no es a que pertenece, es a dado valor de 'x' que valor tendra 'y'.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
xr = np.linspace(0, 6.2, 100)

media, desviacion = 0, 0.1
ruido = np.random.normal(media, desviacion, len(xr))

yr = np.sin(xr) + ruido

df = pd.DataFrame({'xr': xr, 'yr': yr})
sns.scatterplot(data = df, x ='xr', y = 'yr')
```

```python
XXr = xr.reshape(-1, 1)
```

```python
clfr = DecisionTreeRegressor(random_state=99)
clfr.fit(XXr, yr)
```

```python
plt.figure(figsize = (12, 8))
plot_tree(clfr, filled = True, feature_names = xr, class_names = yr)
plt.show()
```

```python
y_predictr = clfr.predict(XXr)
#Aqui vamos a ver nuestro sobreajuste
plt.plot(xr, yr, 'r.', label='datos reales')
plt.plot(xr, y_predictr, 'bx', label='datos observados')
plt.legend()
#hacer matriz de confusion
```

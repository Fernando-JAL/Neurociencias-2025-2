```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r"/home/bubblebeth/GitHub/Neurociencias-2025-2/S03_datasets/iris/iris.csv"

df = pd.read_csv(path)
```

```python
sns.scatterplot(data=df, x='Petal.Length', y = 'Sepal.Length')
```

cada punto es un cluster

```python
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
```

```python
# Normalizando datos
X_std = StandardScaler().fit_transform(df.iloc[:, :-1])
```

# Gaussian Mixture

```python
gm = GaussianMixture(n_components=3, random_state=0).fit(X_std)
```

```python
labels = gm.predict(X_std)
labels
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.scatterplot(data=df, x='Petal.Length', y = 'Sepal.Length', 
                hue='Species', ax=axes[0]).set_title('Iris data')

sns.scatterplot(data=df, x='Petal.Length', y = 'Sepal.Length', 
                hue=labels, palette='deep', ax=axes[1]).set_title('MÃ©todo de Gaussian mixture')
```

```python
df['modelo Gaussiano'] = labels

mapeo = {1: 'setosa', 2: 'virginica', 0: 'versicolor'}
df['modelo Gaussiano'] = df['modelo Gaussiano'] .map(mapeo)
df.head(2)
```

```python
categorias = df.Species.unique()
matriz_conf = {}

for cateogiras_real in categorias:
    real = []
    for categorias_modelo in categorias:
        total = len(df[(df.Species == cateogiras_real) & (df['modelo Gaussiano'] == categorias_modelo)])
        real.append(total)
    matriz_conf[cateogiras_real] = real
matriz_conf
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_gauss = confusion_matrix(df.Species, df['modelo Gaussiano'], labels=df.Species.unique())

sns.heatmap(cm_gauss, annot=True,
           xticklabels = df.Species.unique(), yticklabels = df.Species.unique(),
           cmap = "Blues").set_title("Matriz de confusion de Gaussian mixture")

plt.xlabel('Prediccion')
plt.ylabel("Real")
```

## AHORA CON KMEANS

```python
from sklearn.cluster import KMeans
```

```python
X_std
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_std)
```

```python
df['kmeans'] = kmeans.labels_
df
```

```python
mapeo = {1: 'setosa', 2:'versicolor', 0:'virginica'}
df['kmeans'] = df['kmeans'].map(mapeo)
df
```

```python
cm_kmeans = confusion_matrix(df.Species, df.kmeans, labels = df.Species.unique())

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.heatmap(cm_kmeans, annot=True,
           xticklabels = df.Species.unique(), yticklabels = df.Species.unique(),
           cmap = "Blues", ax = axes[0]).set_title("Matriz de confusion de kmeans")

axes[0].set_xlabel('Prediccion')
axes[0].set_ylabel("Real")

sns.heatmap(cm_gauss, annot=True,
           xticklabels = df.Species.unique(), yticklabels = df.Species.unique(),
           cmap = "Blues", ax = axes[1]).set_title("Matriz de confusion de Gaussian mixture")

axes[1].set_xlabel('Prediccion')
axes[1].set_ylabel("Real")
```

## CLUSTERING JERARQUICO

revisar distacia taxista y distancia euclidiana
EN LINKAGE
Nosotros le vamos a dar nuesteos puntos que son 'y', luego le decimos el metodo = que es la formula single (minimo de dist entre dos clusters) complete (es el maximo) y el average (es el promedio de todas las distancias entre dos clusters), escrita en sting,, metrica tambien en string  

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
Z = linkage(X_std, 'ward')
dendrogram(Z)
plt.plot()
```

los clusters mas juntitos los consideraremos como 1 solo y asi iremos en multiples ocasiones hasta irnos quedando sin clusters. valor de t es distancia

```python
clusters = fcluster(Z, t=10, criterion='distance')
clusters
```

```python
sns.scatterplot(data=df, x='Petal.Length', y = 'Sepal.Length', 
                hue=clusters, palette='deep').set_title('Cluster Jerarquico')
```

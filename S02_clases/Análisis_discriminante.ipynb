```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# Análisis discriminante 

### Lineal 
- Se usa en un conjunto de datos con diferentes categorías ()
- Intenta encontrar las combinaciones lineales de las características que mejor separan estas categorías 
    - encuentra las direcciones que maximizan la la distancie entre las medias (separación) entre clases y minimiza la varianza dentro de cada clase 
- Es útil para la reducción de dimensionalidad cuando tienes muchas características y es más efectivo cuando el número de muestras es mucho mayor que el número de características.

#### 1 Cálculo de medias, varianza y creación del modelo 

```python
iris = load_iris()
X = iris.data[:, [0, 2]]  # Longitud del sépalo y del pétalo
y = iris.target # clases (tipo de flor)

# Dividir en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo LDA
lda = LinearDiscriminantAnalysis()

# **Cálculo de medias y varianzas para cada clase en LDA**
# LDA calcula las medias de cada clase y asume que todas las clases tienen la misma matriz de covarianza.
# Este paso ocurre internamente cuando usamos fit().
lda.fit(X_train, y_train)

```

#### 2 Predicciones y evaluar el modelo 

```python
y_pred = lda.predict(X_test)  # predice sobre los datos de prueba 
accuracy = accuracy_score(y_test, y_pred)  # calcula la precisión del modelo
print(f"Precisión del modelo LDA: {accuracy:.2f}")
```

#### 3 Matriz de confusión 

```python
cm = confusion_matrix(y_test, y_pred)  # Calcula la matriz de confusión
sns.heatmap(cm, annot=True, cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de confusión - LDA")
plt.show()
```

#### 4 Visualizar proyección LDA en 2D 

```python
X_lda = lda.transform(X)  # Proyección de los datos originales en el espacio LDA

plt.figure(figsize=(8, 6))
# Graficamos los puntos proyectados en 2D
for i, target in enumerate(iris.target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1] if X_lda.shape[1] > 1 else [0]*sum(y == i), label=target)
plt.xlabel("LD1")  # Primer componente discriminante
if X_lda.shape[1] > 1:
    plt.ylabel("LD2")  # Segundo componente discriminante
plt.legend()
plt.title("Proyección LDA de los datos Iris")
plt.show()
```

```python
mean = [0, 0]
cov = [[3, 3], [3,4]]
size = 1000
x = np.random.multivariate_normal(mean=mean, cov=cov, size=size)
```

```python
plt.scatter(x[:, 0], x[:,1])
plt.axis('equal')
plt.show()
```

```python
x
```

```python
#normalizar los datos
```

```python
np.mean(x, axis=0), np.std(x, axis=0)
```

```python
x_std =(x -np.mean(x, axis=0))/np.std(x, axis=0)
```

```python
plt.scatter(x_std[:,0], x_std[:,1])
```

```python
# Calculo de la covarianza
```

```python
x_mean = np.mean(x_std, axis = 0)
x_mean
```

```python
cov_matrix = (x_std - x_mean).T.dot((x_std - x_mean))/(x_std.shape[0]-1)
cov_matrix
```

```python
np.cov(x_std.T)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
```

```python
# Calculo de matriz de covarianza, correlacion, valores y vectores propios con la matriz de correlacion
```

```python
corr_matrix = np.corrcoef(x_std.T)
corr_matrix
```

```python
eig_vals_corr, eig_vecs_corr = np.linalg.eig(corr_matrix)
eig_vals_corr, eig_vecs_corr
```

```python
#singular value decomposition
```

```python
u, s, v = np.linalg.svd(x_std.T)
u
```

```python
eig_vals, eig_vecs
```

```python
np.linalg.norm(eig_vecs[0]), np.linalg.norm(eig_vecs[1])
```

```python
np.sum(eig_vals)
```

```python
total_sum = np.sum(eig_vals)

val_porc_eig_val_1= eig_vals[0]/total_sum*100
val_porc_eig_val_2= eig_vals[1]/total_sum*100
val_porc_eig_val_1, val_porc_eig_val_2
```

ordenando valores y vectores propios de mayor a menor

```python
eig_pairs = [(eig_vals[i],eig_vals[i]) for i in range (len(eig_vals))]
eig_pairs
```

```python
eig_pairs.sort(reverse=True)
eig_pairs
```

```python
total_sum = np.sum(eig_vals)

porc_eig = [eig_vals/total_sum*100 for eig_vals, _ in eig_pairs]
cum_porc_eig = np.cumsum(porc_eig)
porc_eig, cum_porc_eig
```

```python
#plots
x=['Val prop 1','Val prop 2']
plt.bar(x, porc_eig)
plt.plot(x,cum_porc_eig, '.-', c='orange', linewidth=5, markersize=15)
```

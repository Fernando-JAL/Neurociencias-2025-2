```python
# Distribucioens multivariadas
# Consideramos dos medias pq etsamos trabajndo en un espacio de 2 dimensiones
```

```python
mean = [0, 0]
cov = [[3, 3], [3, 4]]
size= 500
X = np.random.multivariate_normal(mean=mean, cov=cov, size=size)
X
```

```python
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.show
```

# Normalización de datos 

```python
np.mean(X, axis=0), np.std(X, axis=0)
```

```python
X_std = (X - np.mean(X, axis=0))/np.std(X, axis=0)
```

```python
plt.scatter(X_std[:,0], X_std[:,1])
```

# Calculo de covarianza

```python
X_mean = np.mean(X_std, axis=0)
X_mean
```

```python
cov_matrix = (X_std - X_mean).T.dot((X_std - X_mean))/(X_std.shape[0]-1) #manual
cov_matrix
```

```python
np.cov(X_std.T) # con función lol
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
```

# Cálculo de valores y vectores propios con la matriz de correlacion 

```python
corr_matrix = np.corrcoef(X_std.T)
corr_matrix
```

```python
eig_vals_corr, eig_vecs_corr = np.linalg.eig(corr_matrix)
eig_vals_corr, eig_vecs_corr
```

# Singular vaue decomposition

```python
u, s, v = np.linalg.svd(X_std.T)
u
```

# Análisis en vectores y valores propios

```python
eig_vals, eig_vecs
```

```python
np.linalg.norm(eig_vecs[0]), np.linalg.norm(eig_vecs[1])
```

```python
np.sum(eig_vals)
```

```python
eig_vals
```

```python
total_sum = np.sum(eig_vals)

val_porc_eig_val_1 = eig_vals[0]/total_sum * 100
val_porc_eig_val_2 = eig_vals[1]/total_sum * 100
val_porc_eig_val_1, val_porc_eig_val_2
```

```python
# ordenando valores y vectores propios de mayor a menor
eig_pairs = [(eig_vals[i], eig_vecs[i]) for i in range(len(eig_vals))]
eig_pairs
```

```python
eig_pairs.sort(reverse=True)
eig_pairs
```

```python
total_sum = np.sum(eig_vals)

porc_eig = [eig_val/total_sum * 100 for eig_val, _ in eig_pairs]
cum_porc_eig = np.cumsum(porc_eig)
porc_eig, cum_porc_eig
```

```python
# plots
x = ["val prop 1", "val prop 2"]
plt.bar(x, porc_eig)
plt.plot(x, cum_porc_eig, '.-', c='orange', linewidth=5, markersize=20)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2).fit(X_std)
```

```python
pca.components_, pca.explained_variance_
```

```python
plt.scatter(X_std[:, 0], X_std[:, 1])

for idx, (eig_vec, eig_val) in enumerate(zip(pca.components_, pca.explained_variance_)):
    plt.plot([0, eig_val*eig_vec[0]], [0, eig_val*eig_vec[1]], color=f"C{idx+2}", linewidth=10)
```

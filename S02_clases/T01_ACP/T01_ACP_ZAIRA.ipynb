```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix

# Cargar el conjunto de datos
df = pd.read_csv('/Users/ZvalEnes/Desktop/Neurociencias-2025-2/S03_datasets/iris/iris.csv')

```

```python

# Dividir los datos en caracterÃ­sticas y variable objetivo
X, y = df.iloc[:, :-1], df.iloc[:, -1]
```

```python

# Entrenar un clasificador de Ã¡rbol de decisiÃ³n con una profundidad mÃ¡xima de 3
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=99)
clf.fit(X, y)
```

```python

# Visualizar el Ã¡rbol de decisiÃ³n con colores personalizados
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=df.columns[:-1], class_names=df['Species'].unique())
plt.show()

```

```python

# Predecir los valores observados
observado = clf.predict(X)
```

```python

# Crear una matriz de confusiÃ³n
conf_matrix = confusion_matrix(y, observado)
conf_matrix_df = pd.DataFrame(conf_matrix, index=df['Species'].unique(), columns=df['Species'].unique())
```

```python

# Graficar la matriz de confusiÃ³n
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='terrain')
plt.ylabel('Actual')
plt.xlabel('Predicho')
plt.title('Matriz de ConfusiÃ³n')
plt.show()
```

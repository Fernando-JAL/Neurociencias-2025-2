```python
# distribuciones multivariadas para generar nmeros aleatorios
mean = [0, 0]
cov = [[3, 3], [3, 4]] # matriz de 2x2 pq hay q
size = 500 # generamos mil valores en el eje x, con esos dos valores (la varianza ser esa matriz)
X = np.random.multivariate_normal(mean = mean, cov = cov, size = size)
```

### normalizacin datos

```python
X 
```

```python
np.mean(X, axis = 0), np.std(X, axis = 0)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.axis("equal") # equal para misma escala en eje x, y
plt.show()
```

```python
np.mean(X, axis = 0), np.std(X, axis = 0)
```

```python
X_std = (X - np.mean(X, axis = 0))/np.std(X, axis = 0)
```

### clculo covarianza

```python
X_mean = np.mean(X, axis = 0) # axis = 0 para q nos de un VECTOR
X_mean
```

```python
X - X_mean
```

```python
cov_matrix = (X - X_mean).T.dot((X - X_mean))/(X.shape[0]-1) # .dot para mutliplicacin matricial
```

```python
np.cov(X.T) # esto ya nos saca la covarianza directo
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
```

### clculo de valores y vectores propios con matriz de correlacin

```python
corr_matrix = np.corrcoef(X.T)
corr_matrix
```

```python
eig_vals_corr, eig_vecs_corr = np.linalg.eig(corr_matrix)
eig_vals_corr, eig_vecs_corr
```

```python
np.sum(eig_vals)
```

```python
total_sum = np.sum(eig_vals)
```

```python
valor_porcentual_eig_1 = eig_vals[0]/total_sum * 100
valor_porcentual_eig_1 # el primer eig_vec aporta el 6% de la info
```

```python
valor_porcentual_eig_2 = eig_vals[1]/total_sum * 100
valor_porcentual_eig_2 # el segundo eig_vec aporta el 93% de la info
```

```python
eig_pairs = [(eig_vals[i], eig_vecs[i]) for i in range(len(eig_vals))]
eig_pairs
```

```python
eig_pairs.sort(reverse = True)
eig_pairs
```

```python
total_sum = np.sum(eig_vals)

porc_eig = [eig_val/total_sum * 100 for eig_val, _ in eig_pairs]
cum_porc_eig = np.cumsum(porc_eig)
porc_eig, cum_porc_eig
```

```python
# plots
x = ["valor propio 1", "valor propio 2"]
plt.bar(x, porc_eig)
plt.plot(x, cum_porc_eig, '.-', c = "purple", linewidth = 5, markersize = 20)
```

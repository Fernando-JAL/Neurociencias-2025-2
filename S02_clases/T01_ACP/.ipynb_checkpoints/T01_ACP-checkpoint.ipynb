```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 0, 1
#np.random.normal(mu, sigma)
size = 500
x=np.random.normal(mu, sigma, size=size)
```

```python
plt.hist(x, bins=20)
plt.show()
```

```python
mu_x, sigma_x = 0, 8
mu_y, sigma_y = 0, 2
size = 1000
```

```python
x = np.random.normal(mu_x, sigma_x, size=size)
y = np.random.normal(mu_y, sigma_y, size=size)
```

```python
plt.scatter(x, y)
plt.axis('equal')
plt.show()
```

```python
mean = [0, 0]
cov = [[3, 3], [3,4]]
size = 1000
x = np.random.multivariate_normal(mean=mean, cov=cov, size=size)
```

```python
plt.scatter(x[:, 0], x[:,1])
plt.axis('equal')
plt.show()
```

```python
x
```

```python
#normalizar los datos
```

```python
np.mean(x, axis=0), np.std(x, axis=0)
```

```python
x_std =(x -np.mean(x, axis=0))/np.std(x, axis=0)
```

```python
plt.scatter(x_std[:,0], x_std[:,1])
```

```python
# Calculo de la covarianza
```

```python
x_mean = np.mean(x_std, axis = 0)
x_mean
```

```python
cov_matrix = (x_std - x_mean).T.dot((x_std - x_mean))/(x_std.shape[0]-1)
cov_matrix
```

```python
np.cov(x_std.T)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
```

```python
# Calculo de matriz de covarianza, correlacion, valores y vectores propios con la matriz de correlacion
```

```python
corr_matrix = np.corrcoef(x_std.T)
corr_matrix
```

```python
eig_vals_corr, eig_vecs_corr = np.linalg.eig(corr_matrix)
eig_vals_corr, eig_vecs_corr
```

```python
#singular value decomposition
```

```python
u, s, v = np.linalg.svd(x_std.T)
u
```

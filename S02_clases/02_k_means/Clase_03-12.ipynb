```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_path = r"C:\Users\OMEN CI7\Documents\repository\Neurociencias-2025-2\S03_datasets\iris\iris.csv"
df = pd.read_csv(iris_path)
```

```python
df['Species'].unique()
```

```python
sns.scatterplot(data=df, x='Petal.Length', y='Sepal.Length')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Seleccionamos los datos de interés
X = df.iloc[:, :-1]
```

```python
# Normalizamos data
X_std = StandardScaler().fit_transform(X)
```

# Encontrar cluster por kmeans

```python
# Aplicando kmeans
kmeans = KMeans(n_clusters=3, tol=0.000001)
kmeans.fit(X_std)
```

### Propiedades de los clusters

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

### Creación de dataframe y plots

```python
df_std = pd.DataFrame(data=X_std, columns=df.columns[:-1])
df_std['kmeans'] = kmeans.labels_
```

```python
sns.scatterplot(data=df_std, x='Petal.Length', y='Sepal.Length', hue='kmeans')
sns.scatterplot(x=kmeans.cluster_centers_[:, 3], y=kmeans.cluster_centers_[:, 0], 
                marker="X", color='r', s=300)
```

# Método del codo

```python
inertia = []
for no_clusters in range(2, 15):
    kmeans = KMeans(n_clusters=no_clusters)
    kmeans.fit(X_std)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 15), inertia, '.-')

plt.xlabel('número de clusters')
plt.ylabel('valor de inertia')
plt.title('Método del codo')
```

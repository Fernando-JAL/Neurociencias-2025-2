```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```python
X_std = StandardScaler().fit_transform(df.iloc[:, :-1])

df_std = pd.DataFrame(data=X_std, columns=df.columns[:-1])
df_std['Species'] = df['Species']
```

# K means

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_std)
labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
```

# Método del codo?

```python
score = []
siluette_coef = []

for no_kmeans in range(2, 15):
    gm = GaussianMixture(n_components=3, random_state=0).fit(X_std)
    score.append(gm.score(X_std))
    sil_coeff = silhouette_score(X_std, gm.fit_predict(X_std), metric='euclidean')
    siluette_coef.append(sil_coeff)
```

```python
plt.plot(range(2, 15), score, '.-')

plt.title('Método del codo')
plt.xlabel('Número de kmeans')
plt.ylabel('Valor de inertia')
```

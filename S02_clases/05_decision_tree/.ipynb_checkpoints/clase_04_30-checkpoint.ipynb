```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = r"C:\Users\OMEN CI7\Documents\repository\Neurociencias-2025-2\S03_datasets\iris\iris.csv"

df = pd.read_csv(path)
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
tree = DecisionTreeClassifier(criterion='entropy', random_state=42)
tree.fit(X, y)
```

```python
cv = KFold(n_splits=10, shuffle=True)
```

```python
scores = cross_val_score(tree, X, y, scoring='accuracy', cv=cv)
scores
```

```python
np.mean(scores)
```

```python
for i, (train_index, test_index) in enumerate(cv.split(X)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")
```

# Determinar el número de profundidad para la poda

```python
for i in range(1, 10):
    tree = DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=i)
    tree.fit(X, y)
    
    cv = KFold(n_splits=10, shuffle=True)
    scores = cross_val_score(tree, X, y, scoring="accuracy", cv=cv)
    score = np.mean(scores)
    print(f"La precisión para profundidad = {i} es {score}")
```

# Árbol de decisión para regresión

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
x = np.linspace(0, 6.2, 100)

media, desviacion = 0, 0.1
ruido = np.random.normal(media, desviacion, len(x))

y = np.sin(x)

df = pd.DataFrame({'x': x, 'y': y})
```

```python
X = x.reshape(-1, 1)

tree = DecisionTreeRegressor(max_depth=5)
tree.fit(X, y)
```

```python
plot_tree(tree, filled=True)
plt.show()
```

```python
y_redict = tree.predict(X)

plt.plot(x, y, 'r.', label='datos reales')
plt.plot(x, y_redict, 'bx', label='datos modelo')
plt.legend()
```
